# DATA PREPROCESSING

```python
df.isnull().sum() # Hear no null values so no imputation
```

```python
numerical = df[['Air Pollution', 'Alcohol use',
       'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk',
       'chronic Lung Disease', 'Balanced Diet', 'Obesity', 'Smoking',
       'Passive Smoker', 'Chest Pain', 'Coughing of Blood', 'Fatigue',
       'Weight Loss', 'Shortness of Breath', 'Wheezing',
       'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold',
       'Dry Cough', 'Snoring']]
```

```python
# to finding the outliers by INTERQUATILE RANGE(IRQ) method
for column in numerical:
  q1=df[column].quantile(0.25)
  q3=df[column].quantile(0.75)
  IQR= q3-q1
  lower_bound=q1-1.5*IQR
  upper_bound=q3+1.5*IQR
print(f"Column: Q1: {q1}, Q3: {q3}, IQR: {IQR}, Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```

```python
df=df[(df[column]>=lower_bound) & (df[column]<=upper_bound)]
```

```python
df.shape
```

```python
skewness_columns = df[['Air Pollution', 'Alcohol use',
       'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk',
       'chronic Lung Disease', 'Balanced Diet', 'Obesity', 'Smoking',
       'Passive Smoker', 'Chest Pain', 'Coughing of Blood', 'Fatigue',
       'Weight Loss', 'Shortness of Breath', 'Wheezing',
       'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold',
       'Dry Cough', 'Snoring']]
```

```python
# finding the skewness score for disturbution of the data
from scipy.stats import skew
```

```python
skewness = skewness_columns.skew() # skewness score
print(skewness)
```

```python
import numpy as np
skewed_columns = skewness[skewness.abs() > 0.5].index # squre root transformation for > 0.5 sk score
for column in skewed_columns:
    df[column] = np.sqrt(df[column])
```

```python
df.head()
```

```python
new_skewness = skewness_columns.skew()
```

```python
new_skewness
```

```python
negatively_skewed = skewness[skewness < 0].index
```

```python
for column in negatively_skewed: # taking the negative skew to transform
    df[column] = np.sqrt(df[column] - df[column].min() + 1)
```

# Feature Selection

```python
X = df.drop(columns=['Patient Id','Age','Gender', 'Level'])  # Features columns
y = df['Level']    # Target columns
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
k=15
selector = SelectKBest(score_func=f_classif,k=k)
selector.fit(X, y)
# By using the select k best method to find the selected best futures for analysis
```

```python
select_features= X.columns[selector.get_support()]
print(select_features) # placeing in x.columns as selected features
```

```python
X_selected = X[select_features]
```

```python
scores = selector.scores_
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': scores}).sort_values(by='Score', ascending=False)
feature_scores
```

# Build the Machine Learning Model

```python
from sklearn.linear_model import LogisticRegression # importing all the ml unsupervised model to  find the best model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
classifiers = {
    'Logistic Regression': LogisticRegression(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Classifier': SVC(),
    'Naive Bayes': GaussianNB(),
    'XGBoost': XGBClassifier()
}
```

```python
from sklearn.metrics import accuracy_score, f1_score # importing the evaluvate metrix  accuracy score fl score
results = {} # logistic regression and svc suppot vector classifier are the best most performed model
for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred,average= 'weighted')
    results[name] = {'Accuracy': accuracy, 'F1 Score': f1}
results_df = pd.DataFrame(results).T
results_df
```
